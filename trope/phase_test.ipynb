```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io.wavfile import write
```

```python
sample_rate = 22050
```

```python
# phase = 0
# figure out the phase increment (the same as tone with the (each_)sample factored out)
# phase_increment = (2 * np.pi) * frequency / sample_rate
# update a variable representing the current phase? 


def generate_sine(frequency, duration, amplitude=0.5):
    each_sample = np.arange(int(duration * sample_rate))

    tone = amplitude * np.sin((2 * np.pi) * each_sample * frequency / sample_rate)
    
    return tone
```

```python
tone_220 = generate_sine(80, 0.25)
tone_220[-50:]
```

```python
idx = np.nonzero(np.isclose(tone_220, 0.0, rtol=1e-1))

print(idx)
last_close_to_zero = idx[0][-1]
length_of_concern = len(tone_220) - last_close_to_zero
print(f'{length_of_concern / (1.2*22050) * 1000 = }')
print(f'{last_close_to_zero = }')
```
